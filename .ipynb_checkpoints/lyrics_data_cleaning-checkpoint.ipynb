```python
# Import dependencies
# spotipy is a library with methods that call the Spotify API

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
# Establish connection to Spotify

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
#Load songs csv

songs_df = pd.read_csv("songlyrics/songdata.csv")
songs_df.head()
```

```python
#Remove newline \n characters from lyric text
songs_df['text'].replace('\\n','', regex=True, inplace=True)

#Remove extra whitespace from text column
songs_df['text'].replace('   ', ' ', regex=True, inplace=True)

#Overwrite existing csv to remove newlines
songs_df.to_csv("songlyrics/songdata.csv", index=False)

songs_df.head()
```

```python
# Empty arrays to store data
artist_name = []
track_name = []
track_id = []


# Nested loop to call Spotify API search endpoint for each track
for i in songs_df.itertuples():
    track_results = sp.search(q="track:" + i.song + " artist:" + i.artist, type='track', limit=1)
    #print(f"data {i.Index} completed")
    
    # Identify tracks that were not found on Spotify
    if len(track_results['tracks']['items']) == 0:
    
        track_name.append("Track not found")
        track_id.append("Track not found")
        
    else:
        
        # Get artist name, track name and track id for songs that are found in search
        for i, t in enumerate(track_results['tracks']['items']):
            artist_name.append(t['artists'][0]['name'])
            track_name.append(t['name'])
            track_id.append(t['id'])
```

```python
# Add track id to dataframe

songs_df["Spotify track id"] = track_id
songs_df = songs_df[["artist", "song", "text", "Spotify track id"]]
songs_df.head()
```

```python
# Use .loc to remove songs with no track id 

song_id_df = songs_df.loc[songs_df["Spotify track id"] != "Track not found"]
song_id_df.head()
```

```python
# Save to csv to prevent long loop 
song_id_df.to_csv("songs_with_ids.csv", index=False)
```

```python
# Reload csv instead of running loop again 
song_id_df = pd.read_csv("songs_with_ids.csv")

# Remove triple spaces from text column
song_id_df['text'].replace('   ', ' ', regex=True, inplace=True)

# Save over csv
song_id_df.to_csv("songs_with_ids.csv", index=False)

song_id_df.head()
```
